Objective

Develop basic recommendation systems using Python and pandas.
Suggesting movies that are most similar to a particular movie.



```python
import pandas as pd
```

```python
user_df = pd.read_csv('user.data',sep='\t')
user_df.columns = ['user_id','item_id','rating','timestamp']
user_df.head()
```

```python
user_df.info()
```

```python
movie_df = pd.read_csv('Movie_Titles.txt')
movie_df.head()
```

```python
movie_df.info()
```

Lets merge both data frames on ‘item_id’ which is present in both data frames.

```python
df = pd.merge(user_df,movie_df,on='item_id')
df.head()
```

Grouping movies here and taking the mean of all ratings given to them and then we are sorting them by their mean rating.

```python
rating_and_no_of_rating = pd.DataFrame(df.groupby('title')['rating'].mean().sort_values(ascending=False))
rating_and_no_of_rating
```

Lets add a column named of number_of_ratings (which represents number of people who rated the movie)

```python
rating_and_no_of_rating['number_of_ratings'] = df.groupby('title')['rating'].count()
rating_and_no_of_rating
```

Lets sort by number of ratings to get geniue ratings. In the above case movie ‘They made a criminal’ is showing up which might has got a rating from only one person and they rated it 5 stars. That’s why its mean is also 5.

```python
rating_and_no_of_rating = rating_and_no_of_rating.sort_values('number_of_ratings',ascending=False)
rating_and_no_of_rating.head()
```

Now that we have a movies with geniue rating, lets find the the user who have watch these movies.


Let’s create a matrix that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. 

```python
pt = df.pivot_table(index='user_id',columns='title',values='rating')
pt.head()
```

In the above pivot table, NaN represents that, that user has not watched that movie.

```python
test_movie = input('Enter movie name --> ')

movie_vector = pt[test_movie].dropna()
similar_movies = pt.corrwith(movie_vector)

corr_df = pd.DataFrame(similar_movies,columns=['Correlation'])
corr_df = corr_df.join(rating_and_no_of_rating['number_of_ratings'])

corr_df = corr_df[corr_df['number_of_ratings']>100].sort_values('Correlation',ascending=False).dropna()
corr_df.head(10)
```
