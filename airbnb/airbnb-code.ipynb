```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('airbnb.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.drop('last_review',axis=1,inplace=True)
```

```python
df.drop('reviews_per_month',axis=1,inplace=True)
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df.drop('host_name',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(11,8))
sns.countplot(data=df,x='neighbourhood_group',palette='Oranges')
plt.title('Number of rooms within neighbourhood group\n')
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.countplot(data=df,x='neighbourhood_group',hue='room_type',palette='Oranges')
plt.title('Number of room types within neighbourhood group \n')
plt.show()
```

```python
plt.figure(figsize=(11,8))
sns.countplot(x='room_type',data=df,palette='Oranges')
plt.title('Most Preferred Rooms by Customers')
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.violinplot(data=df[df.price <500], x='neighbourhood_group', y='price', palette='GnBu_d')
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.violinplot(data=df[df.minimum_nights <40], x='neighbourhood_group', y='minimum_nights', palette='GnBu_d')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='neighbourhood_group',palette='Oranges')
plt.show()
```
